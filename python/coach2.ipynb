```python
class MyCoach2:
    """
    鬼コーチクラス
    """
    count = 0
    max = 5     #上限を設定するクラス変数
    
    def teach(self):
        """熱血指導メソッド
        """
        if self.__class__.count < self.__class__.max:
            # countがmax未満のときの処理
            print('もっと強く‼')
        else:
            print('よーしOKだ')
        self.__class__.count += 1
```

```python
my1 = MyCoach2()
for i in range(3):
    print('スマッシュ')
    my1.teach()
    

my2 = MyCoach2()
for i in range(3):
    print('ライジング')
    my2.teach()
```

```python
oni1 = MyCoach2()
for i in range(3):
    print('スマッシュ')
    oni1.teach()
oni2 = MyCoach2()
for i in range(3):
    print('ライジング')
    oni2.teach()
MyCoach2.count = 0
oni3 = MyCoach2()
for i in range(2):
    print('トップスピン')
    oni3.teach()
```
