```python
class MyCoach:
    """鬼コーチクラス"""

    def __init__(self, max):
        """初期化メソッド

        Args:
            max (int):      繰り返しの回数

            """
        self.max = max
        self.count = 0

    def teach(self):
        """熱血指導メソッド
        """

        if self.count < self.max:
            # self.countがself.max未満の時の処理
            print('もっと強く!')

        else:
            print('よーしOKだ!')
        
        self.count += 1
```

```python
mc = MyCoach(5)
for i in range(6):
    print('スマッシュ')
    mc.teach()
```
